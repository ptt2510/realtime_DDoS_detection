```python
from kafka import KafkaAdminClient, KafkaConsumer, KafkaProducer
import sys
import json
from json import loads
import numpy as np
import requests
import joblib
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler(feature_range=(0, 1))
### Setting up the Python consumer
bootstrap_servers = ['localhost:9092']
topicName = 'DDOS3'
consumer = KafkaConsumer (topicName,bootstrap_servers = bootstrap_servers,
    auto_offset_reset = 'earliest',value_deserializer=lambda x: loads(x.decode('utf-8')))  ## You can also set it as latest
dict={
    0 : 'BENIGN',
    1 : 'UDP',
    2 : 'Syn',
    3 : 'MSSQL',
    4 : 'NetBIOS',
    5: 'LDAP',
    6 : 'Portmap'
}  
sm=joblib.load(r"C:\Users\Admin\Downloads\softmax.pkl")
### Reading the message from consumer
try:
    for message in consumer:
        arr=[]
        t=message.value
        t=t.replace('[','')
        t=t.replace(']','')
        for i in t.split(','):
            arr.append(float(i))
        data=np.array(arr)
        data = data.reshape(1, 73)
        data=scaler.fit_transform(data)
        y_pred = sm.predict(data)
        kq=int(y_pred)
        mess=dict[kq]
        if int(y_pred)!=0:
            payload='{"text":"%s"}' % mess
            response=requests.post('https://hooks.slack.com/services/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',
                              data=payload)
            print(response.text)
        else:
            print(mess)
except KeyboardInterrupt:
    sys.exit()
    
```
